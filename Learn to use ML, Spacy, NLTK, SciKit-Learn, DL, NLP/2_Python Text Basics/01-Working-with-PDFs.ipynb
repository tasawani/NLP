```python
# Function to write a string to a .txt file
def write_to_txt(function_name, filename):
    try:
        # Write the function name (parameter) to a .txt file
        with open(filename, 'w', encoding='utf-8') as file:
            file.write(function_name)
        print(f"Function name '{function_name}' successfully written to {filename}")
    
    except Exception as e:
        print(f"An error occurred: {e}")
```

```python
# Example usage:
value = page_to_text
filename = "menu.txt"
write_to_txt(value, filename)
```
